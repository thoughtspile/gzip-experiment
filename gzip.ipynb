```python
import pandas as pd
```

```python
df = pd.read_csv('css-gzip.tsv', sep='\t', names=['gzip_css'])
js = pd.read_csv('js-gzip.tsv', sep='\t', names=['gzip_js'])
svg = pd.read_csv('svg-gzip.tsv', sep='\t', names=['gzip_svg'])
mixed = pd.read_csv('mixed-gzip.tsv', sep='\t', names=['gzip_mixed'])
df = df.join(js).join(svg).join(mixed)
df['raw'] = df.index
```

```python
df[['gzip_js', 'gzip_css', 'gzip_svg', 'gzip_mixed']].plot()
```

```python
df['js_ratio'] = (df['raw'] / df['gzip_js'])
df['css_ratio'] = (df['raw'] / df['gzip_css'])
df['svg_ratio'] = (df['raw'] / df['gzip_svg'])
df['mixed_ratio'] = (df['raw'] / df['gzip_mixed'])
df[['js_ratio', 'css_ratio', 'svg_ratio', 'mixed_ratio']].plot(figsize=(15,3), ylim=0, xticks=[10910, 30758], grid=True)
```
